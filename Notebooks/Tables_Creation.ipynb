## Product Table

```python
# Step 1: Get unique elements in product_ID column
unique_product_ids = df['Product.ID'].unique()

# Step 2: Create the new DataFrame
new_rows = []

for product_id in unique_product_ids:
    # Get rows for the current product ID
    product_rows = df[df['Product.ID'] == product_id].iloc[0]  # Take the first row
    
    # Calculate new price
    new_price = product_rows['Sales'] / product_rows['Quantity']
    
    # Generate random size
    random_size = np.random.uniform(1.0, 150.0)  # Change to float
    
    # Append to the new rows list
    new_rows.append({
        'Product_ID': product_id,
        'Product_Name': product_rows['Product.Name'],
        'Category': product_rows['Category'],
        'Sub_Category': product_rows['Sub.Category'],
        'Unit_Price': new_price,
        'Size': random_size
    })

# Create the new DataFrame
new_df = pd.DataFrame(new_rows)
```

## Customers Table

```python
import random
import string

def generate_random_email(domain='example.com'):
  """Generate a random email address without commas."""
  username_length = random.randint(5, 10)
  # Use filter to remove commas from the character pool
  allowed_chars = filter(lambda char: char != ",", string.ascii_lowercase + string.digits)
  # Convert the filter object to a list
  allowed_chars_list = list(allowed_chars)
  username = ''.join(random.choices(allowed_chars_list, k=username_length))
  return f"{username}@{domain}"

def generate_random_password(length=12):
  """Generate a random password without commas."""
  characters = string.ascii_letters + string.digits + string.punctuation.replace(",", "")
  password = ''.join(random.choices(characters, k=length))
  return password
    
# Generate a random email and password
random_email = generate_random_email('gmail.com')
random_password = generate_random_password()

print(f"Random Email: {random_email}")
print(f"Random Password: {random_password}")
```

```python
# Step 1: Get unique elements in product_ID column
unique_customers_ids = df['Customer.ID'].unique()

# Step 2: Create the new DataFrame
new_rows2 = []

for customer_id in unique_customers_ids:

    customer_rows = df[df['Customer.ID'] == customer_id].iloc[0]  # Take the first row
        
    # Append to the new rows list
    new_rows2.append({
        'Customer_ID': customer_id,
        'Customer_Name': customer_rows['Customer.Name'],
        'Segment': customer_rows['Segment'],
        'Email': generate_random_email('gmail.com'),
        'Password': generate_random_password()
    })

# Create the new DataFrame
new_df2 = pd.DataFrame(new_rows2)
```

## Customer_Orders Table

```python
def randomize_status():
    """Randomly select a status from the list."""
    statuses = ["shipped", "in progress", "pending"]
    return random.choice(statuses)

# Example usage
random_status = randomize_status()
print(f"Random Status: {random_status}")
```

```python
# Step 1: Get unique elements in product_ID column
unique_customers_orders_ids = df['Order.ID'].unique()

# Step 2: Create the new DataFrame
new_rows3 = []

for order_id in unique_customers_orders_ids:

    order_rows = df[df['Order.ID'] == order_id].iloc[0]  # Take the first row
        
    # Append to the new rows list
    new_rows3.append({
        
        'Order_ID': order_id,
        'Order_Date': order_rows['Order.Date'],
        'Order_Priority': order_rows['Order.Priority'],
        'Customer_ID': order_rows['Customer.ID'],
        'Order_Status': randomize_status()
    })

# Create the new DataFrame
new_df3 = pd.DataFrame(new_rows3)
```
